```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Type of Plots

## Line Plots

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
plt.plot(x, y)
plt.title("Simple Line Plot");
plt.xlabel('x-axis label');
plt.ylabel('y-axis label');
plt.grid(True);
```

## Bar Chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [100, 200, 300, 500]
plt.figure(figsize=(10, 6));
plt.bar(months, revenue);
plt.title("Monthly Revenue Bar Chart");
plt.xlabel('Months');
plt.ylabel('Revenue');
```

## Histrogram

```python
marks50 = np.random.randint(0, 100, 50)
```

```python
marks50
```

```python
plt.hist(marks50);
plt.xlabel("Marks");
plt.ylabel("Number of students");
```

## Scatter plot

```python
heights = [65, 70, 68, 72, 66]
weights = [150, 160, 155, 180, 145]
plt.scatter(heights, weights);
plt.title("Heights VS Weights");
plt.xlabel("Heights (inches)");
plt.ylabel("Weights (pounds)");
```

## Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20,25, 30, 40]
predicted_price = [10, 21, 25, 28, 30]
plt.plot(days, predicted_price, 'b-o', label="Original Prices");
plt.plot(days, original_price, 'r-o', label="Predicted Price");
plt.legend();
plt.title("Plot with legends");
plt.xlabel("Weekdays");
plt.ylabel("Price");
```
